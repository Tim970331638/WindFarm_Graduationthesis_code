```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl
mpl.rcParams['font.family']='SimHei'# 设置中文黑体
mpl.rcParams['axes.unicode_minus']= False # 解决负号显示问题

# 读取数据
data = pd.read_csv("../数据集/std_data.csv")
df = pd.DataFrame(data)
std_realpower = df["std_realpower"]

# 画直方图
plt.hist(std_realpower, bins=40, range=(0, 1), density=True, alpha=0.5, edgecolor='black', label='实际值频率柱状图')  # 使用 density=True 将直方图归一化为概率密度

plt.xlabel('std_realpower')
plt.ylabel('freq')

# 添加KDE密度估计曲线
sns.kdeplot(std_realpower, color='red', linestyle='-', label='KDE')

plt.legend()
plt.grid(False)  # 去掉网格
plt.show()
```
