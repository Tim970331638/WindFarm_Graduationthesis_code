# 功率波动平抑测算（可视化）

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
path = '../数据集/'
PReal = pd.DataFrame(pd.read_csv(path+'std_Pc_thy.csv'))["averages"]
Pc_thy = pd.DataFrame(pd.read_csv(path+'std_Pc_thy.csv'))["Pc_thy"]
# 使用 np.where 来将小于或等于 1e-5 的值置为零
Pc_thy = np.where(abs(Pc_thy) > 1e-5, Pc_thy, 0)


# 读取数据
df = pd.read_csv(path+'std_data_5min.csv')

# 计算 ad_real
ad_real = PReal + Pc_thy

def draw_aday(df, date='2022-01-01'):
    # 将 'Datetime' 列转换为 datetime 类型
    df['Datetime'] = pd.to_datetime(df['Datetime'])

    # 从DataFrame中选择特定一天的数据
    selected_day_data = df[df['Datetime'].dt.date == pd.to_datetime(date).date()]

    # 创建交互式图表
    fig = go.Figure()

    # 添加 ad_real
    fig.add_trace(go.Scatter(x=selected_day_data.index,
                             y=ad_real[selected_day_data.index],
                             mode='lines',
                             name='ad_real',
                             line=dict(color='black', dash='dash')))

    # 判断并添加 std_realpower
    if 'std_realpower' in selected_day_data.columns:
        fig.add_trace(go.Scatter(x=selected_day_data.index,
                                 y=selected_day_data['std_realpower'],
                                 mode='lines',
                                 name='std_realpower'))

    # 添加布局，设置图例在图中，并调整字体大小
    fig.update_layout(title='{}数据可视化(测算)'.format(date),
                      xaxis_title='索引',
                      yaxis_title='值',
                      legend=dict(
                          x=0.02,  # 控制图例在x轴上的位置
                          y=0.98,  # 控制图例在y轴上的位置
                          font=dict(
                              size=14  # 设置图例字体大小为14
                          )
                      ))

    # 显示图表
    fig.show()

# 调用函数绘制图表
draw_aday(df, date='2022-1-1')

```
