# 可视化三个时间尺度的预测数据、实际值数据、功率因数（5min尺度）

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import plotly.graph_objs as go
path = '../数据集/'
def draw_aday(df,date = '2022-01-01'):
    # 将 'Datetime' 列转换为 datetime 类型
    df['Datetime'] = pd.to_datetime(df['Datetime'])

    # 从DataFrame中选择特定一天的数据
    selected_day_data = df[df['Datetime'].dt.date == pd.to_datetime(date).date()]

    # 创建交互式图表
    fig = go.Figure()

    # 添加线条
    for column in selected_day_data.columns[1:]:
        fig.add_trace(go.Scatter(x=selected_day_data['Datetime'],
                                y=selected_day_data[column],
                                mode='lines',
                                name=column))

    # 添加布局
    fig.update_layout(title='{}数据可视化'.format(date),
                    xaxis_title='时间',
                    yaxis_title='值')

    # 显示图表
    fig.show()
df = pd.DataFrame(pd.read_csv(path+'std_data_5min.csv'))

draw_aday(df,date = '2022-02-04')
```
