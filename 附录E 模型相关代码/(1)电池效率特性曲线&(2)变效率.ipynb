```python
import numpy as np
import matplotlib.pyplot as plt

# Pc
Pc = np.linspace(-1, 1, 100)

# eta
etas = np.linspace(1, 0.5, 10)

# 
Delta_t = 1

# 0

# 
plt.figure(figsize=(10, 6))

# 
color_map = plt.cm.get_cmap('viridis', len(etas))

# etaDeltaSOC
for i, eta in enumerate(etas):
    Delta_SOC = ((np.maximum(-Pc, 0) * eta) - (np.maximum(Pc, 0) / eta)) * Delta_t
    plt.plot(Pc, Delta_SOC, label=f'eta = {eta:.2f}', color=color_map(i))

# 
plt.legend()

# 
plt.xlabel('Pc')
plt.ylabel('Delta SOC')

# 
plt.title('Battery Charge/Discharge Efficiency Characteristics')

# 0
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# 
plt.grid(True)
plt.show()
```

```python
# PcPc
Pc_positive_linear = np.linspace(0, 1, 100)

# 
# Pc=0=0.98Pc=1=0.5
#  y = mx + b m = (y2 - y1) / (x2 - x1)
m = (0.5 - 0.98) / (1 - 0)
b = 0.98
eta_positive_linear = m * Pc_positive_linear + b

# DeltaSOC
Delta_SOC_positive_linear = ((np.maximum(-Pc_positive_linear, 0) * eta_positive_linear) - (np.maximum(Pc_positive_linear, 0) / eta_positive_linear)) * Delta_t

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(Pc_positive_linear, Delta_SOC_positive_linear, label='Linear eta (Pc > 0)')

# 
plt.legend()

# 
plt.xlabel('Pc (Positive)')
plt.ylabel('Delta SOC')

# 
plt.title('Linear Efficiency Characteristics for Lithium Battery (Pc > 0)')

# 0
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# 
plt.grid(True)
plt.show()
```
