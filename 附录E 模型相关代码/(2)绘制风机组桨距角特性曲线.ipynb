```python
import numpy as np
import matplotlib.pyplot as plt

# 
c1 = 0.5176
c2 = 116
c3 = 0.4
c4 = 5
c5 = 21
c6 = 0.0068

# lambda
lambdas = np.linspace(0, 15, 100)

# beta
betas = np.arange(0, 21, 5)

# 
plt.figure(figsize=(10, 6))

# 
color_map = plt.cm.get_cmap('viridis', len(betas))

# betac_p
for i, beta in enumerate(betas):
    lambda_i_inv = 1 / (lambdas + 0.08 * beta) - 0.035 / (beta**3 + 1)
    lambda_i = 1 / lambda_i_inv
    c_p = c1 * (c2 / lambda_i - c3 * beta - c4) * np.exp(-c5 / lambda_i) + c6 * lambdas
    plt.plot(lambdas, c_p, label=f'beta = {beta} deg', color=color_map(i))

# 
plt.legend()

# 
plt.xlabel('Lambda')
plt.ylabel('Cp')

# 
plt.title('Cp-Lambda Characteristics for Different Pitch Angles (Beta)')

# 
plt.grid(True)
plt.show()
```
